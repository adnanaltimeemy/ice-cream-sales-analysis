```python
# Cell 1: imports
import pandas as pd
import numpy as np
```

```python
# Cell 2: example sales and profit dataset
sales_profit_data = pd.DataFrame({
    "Year": [2022, 2023, 2024],
    "Region": ["Global", "Global", "Global"],
    # Sales in millions of euros; 2024 uses the reported ~1100 as an example.
    "Sales_EUR_million": [950, 1025, 1100],
    # Profit margin as a percent; placeholder estimates.
    # You can adjust based on your research or modeling needs.
    "Profit_Margin_pct": [12.0, 13.5, 14.5]
})

sales_profit_data
```

```python
# Cell 3: demographics dataset example
demographics_uk_2025 = pd.DataFrame({
    "Market": ["UK"],
    "Year": [2025],
    "Pct_Age_18_34": [44.0],
    "Pct_Age_35_54": [43.0],
    "Pct_Age_55_plus": [13.0],
    "Pct_Male": [55.0],
    # If you add female %, could compute 100 - male.
    "Pct_Female": [45.0]
})

demographics_uk_2025
```

```python
# Cell 4: simple merge example
# Suppose we want to attach UK demographics to the latest global data

latest_sales = sales_profit_data.loc[sales_profit_data["Year"] == 2024].copy()
latest_sales["Market"] = "UK"  # example assumption for merge

merged = latest_sales.merge(demographics_uk_2025, on="Market", how="left")

merged
```

```python
# Cell 5: save to CSV
sales_profit_data.to_csv("ben_jerrys_sales_profit.csv", index=False)
demographics_uk_2025.to_csv("ben_jerrys_demographics_uk_2025.csv", index=False)
merged.to_csv("ben_jerrys_sales_with_demo.csv", index=False)
```

```python
# Cell 6: plot trend (requires matplotlib)
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.plot(sales_profit_data["Year"], sales_profit_data["Sales_EUR_million"], marker='o')
plt.title("Ben & Jerry's Sales Trend")
plt.xlabel("Year")
plt.ylabel("Sales (EUR million)")
plt.grid(True)
plt.show()

plt.figure(figsize=(6,4))
plt.plot(sales_profit_data["Year"], sales_profit_data["Profit_Margin_pct"], marker='o', color='orange')
plt.title("Profit Margin Trend")
plt.xlabel("Year")
plt.ylabel("Profit Margin (%)")
plt.grid(True)
plt.show()
```

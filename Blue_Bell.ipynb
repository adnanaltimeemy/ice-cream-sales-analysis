```python
# Step 1: Import libraries
import pandas as pd
import numpy as np

# Step 2: Define regions and years
regions = ["Texas", "Louisiana", "Alabama", "Florida", "Georgia", "South Carolina"]
years = list(range(2018, 2026))

# Step 3: Generate synthetic data
data = []
np.random.seed(42)

for year in years:
    for region in regions:
        total_sales = np.random.uniform(20_000_000, 200_000_000)  # Sales in USD
        profit_margin = np.random.uniform(5, 25)  # Profit margin in %
        avg_age = np.random.uniform(30, 55)  # Customer avg age
        median_income = np.random.uniform(40_000, 85_000)  # Median household income
        percent_frequent = np.random.uniform(10, 40)  # % frequent buyers

        data.append({
            "Year": year,
            "Region": region,
            "Total_Sales_USD": round(total_sales, 2),
            "Profit_Margin_pct": round(profit_margin, 2),
            "Avg_Age": round(avg_age, 1),
            "Median_Income": round(median_income, 0),
            "Percent_Frequent_Buyers": round(percent_frequent, 1)
        })

# Step 4: Build DataFrame
bb_df = pd.DataFrame(data)

# Step 5: Save to CSV
csv_file_path = "blue_bell_sales_demographics.csv"
bb_df.to_csv(csv_file_path, index=False)

# Step 6 (optional): Display sample
bb_df.head(10)
```
